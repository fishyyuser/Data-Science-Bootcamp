```python
from sklearn.datasets import make_classification
import pandas as pd
```

```python
x,y=make_classification(n_samples=1000,n_redundant=0,
n_features=2,n_clusters_per_class=1,
weights=[0.9],random_state=12)
```

```python
df1=pd.DataFrame(x,columns=['feature 1','feature 2'])
df2=pd.DataFrame(y,columns=['target'])
```

```python
df=pd.concat([df1,df2],axis=1)
df.head()
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['feature 1'],df['feature 2'],c=df['target'],s=10,cmap='viridis')
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample=SMOTE()
x,y=oversample.fit_resample(df[['feature 1','feature 2']],df['target'])
```
