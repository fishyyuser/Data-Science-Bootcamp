```python
import pandas as pd
import numpy as np
```

```python
corpus="""Hello My name is Pradyumn Bisht.
I am currently learning NLP from Krish's Videos.
"""
```

```python
print(corpus)
```

```python
#Tokkenizastion
# Corpus(paragraph) -> Documents(sentences)
from nltk.tokenize import sent_tokenize
```

```python
documents=sent_tokenize(corpus)
for document in documents:
    print(document)
```

```python
### Tokenisation
### Corpus -> Words
### documents -> Words
from nltk.tokenize import word_tokenize
```

```python
words=word_tokenize(corpus)
for word in words:
    print(word)
print(f"The length of this words : {len(words)}")
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
```

```python
words=tokenizer.tokenize(corpus)
for word in words:
    print(word)
print(f"The length of this words : {len(words)}")
```
